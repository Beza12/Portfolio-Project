#### Extract the 'Day of the week' from the Date  column

```python
df_concatenated['Date'][0]
```

```python
df_concatenated['Date'][9].day
```

```python
df_concatenated['Date'][0].weekday()
```

```python
# to see on which day the employees mostly be absent
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_concatenated['Day of the week']= df_concatenated['Date'].apply(date_to_weekday)
df_concatenated.head()
```

```python
df_concatenated['Education'] = df_concatenated['Education'].map({1:0, 2:1, 3:1, 4:1})
```
