```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
used_car = pd.read_csv('bar_chart_data.csv')
used_car
```

```python
plt.bar(x= used_car['Brand'], height= used_car['Cars Listings'], color = 'midnightblue')
#plt.figure(figsize=(9,6))
plt.xticks(rotation =40)
plt.yticks(fontsize = 13)
plt.ylabel('Number of Listing',fontsize=13)
plt.title('Car Listing', fontsize =16, fontweight= 'bold')
plt.show()
plt.savefig('Used Car Bar.png')
```

### Pie Chart

```python
fuel_engine_type = pd.read_csv('pie_chart_data.csv')
fuel_engine_type
```

```python
plt.figure(figsize = (7, 6))
plt.pie(fuel_engine_type['Number of Cars'], 
        labels = fuel_engine_type['Engine Fuel Type'].values, 
        autopct = '%.2f%%',
        textprops ={'size':'x-large','fontweight':'bold','color':'w','rotation':35})
plt.legend(loc ='upper left')
plt.title('Cars by Fuel Type', fontweight= 'bold')
plt.show()
```

### Area Chart

```python
fuel_engine_types = pd.read_csv('stacked_area_chart_data.csv')
fuel_engine_types
```

```python
labels = ['Diesel','Petrol','Gas']
plt.figure(figsize=(10,4))
plt.stackplot(fuel_engine_types['Year'],
              fuel_engine_types['Diesel'],
              fuel_engine_types['Petrol'],
              fuel_engine_types['Gas'],
            edgecolor='none')
plt.xticks(fuel_engine_types['Year'], rotation = 45)
plt.ylabel('Number of cars', fontsize= 13)
plt.title('Population of Engine Fuel Types(1982-2016)', fontsize =14, fontweight='bold')
plt.legend(labels = labels, loc = 'upper left')
plt.show()
```

### Line Chart


```python
df_line_chart= pd.read_csv('line_chart_data.csv')
df_line_chart
```

```python
df_line_chart['new_date']=pd.to_datetime(df_line_chart['Date'])
```

```python
df_line_chart['new_date']
```

### Scatter Plot

```python
df_scatter= pd.read_csv('scatter_data.csv')
df_scatter
```

```python
plt.figure(figsize = (12,8))
scatter = plt.scatter(df_scatter['Area (ft.)'],df_scatter['Price'],
                      alpha =0.6,
                      c = df_scatter['Building Type'],
                      cmap = 'mako')
plt.legend(*scatter.legend_elements(),loc ='upper right', title = 'Building Type')
plt.title('Relation between Area and Price', fontsize=14, fontweight= 'bold')
plt.ylabel("Price(000's of $)", weight= 'bold')
plt.xlabel('Area (sq.ft)',weight= 'bold' )
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(df_scatter['Area (ft.)'],
                df_scatter['Price'],
                hue = df_scatter['Building Type'],
                palette = ['black', 'darkblue','purple', 'pink', ' white'],
                s= 100)
                
plt.title('Relation between Area and Price', fontsize=14, fontweight= 'bold')
plt.ylabel("Price(000's of $)", weight= 'bold')
plt.xlabel('Area (sq.ft)',weight= 'bold' )
plt.show()
```

## Regression Scatter Plot

```python
df_reg_scatter = pd.read_csv('scatter_plot_ii.csv')
df_reg_scatter
```

```python
plt.figure(figsize = (8,6))
sns.regplot(x = 'Budget', y = 'Sales', data =df_reg_scatter,
           scatter_kws = {'color' :'k'},
           line_kws = {'color' : 'red'})
plt.title('Effect of AD on Sales', fontsize=14, fontweight= 'bold')
plt.ylabel("Sales in(000's Units)", weight= 'bold')
plt.xlabel("Ad Expenditure in (000's $)",weight= 'bold' )
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.lmplot(x = 'Budget', y = 'Sales', data =df_reg_scatter,
          height = 10,
          scatter_kws = {'color':'k'},
          line_kws = {'color': 'red'})
plt.title('Effect of AD on Sales', fontsize=14, fontweight= 'bold')
plt.ylabel("Sales in(000's Units)", weight= 'bold')
plt.xlabel("Ad Expenditure in (000's $)",weight= 'bold' )
plt.show()
```

## Combinatiion Chart

```python
from matplotlib.ticker import PercentFormatter
df_dual_chart = pd.read_csv('bar_line_chart_data.csv')
df_dual_chart
```

```python
fig, ax =plt.subplots(figsize = (7,5))

ax.bar(df_dual_chart['Year'],
       df_dual_chart['Participants'],
       color = 'k')
ax.set_ylabel('Number of Participants', weight= 'bold')
ax.tick_params(axis = 'y',width = 2, labelsize= 'large')
ax1= ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))
ax1.plot(df_dual_chart['Year'],
       df_dual_chart['Python Users'],
       color = 'red',
        marker = 'D')
ax1.set_ylabel('Python Users', weight= 'bold',color = 'red')
ax1.tick_params(axis = 'y',color = 'red',width = 2, labelsize= 'large')
```
